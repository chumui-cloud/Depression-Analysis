```python
import pandas as pd
import numpy as np
import datetime
import random
```

```python
def train(dataframe):
    df1=dataframe.copy()
    class_status = df1['How are you feeling right now?']
    class_scale = df1['On a scale of 1-100, how would you express this feeling?']
    df = df1.drop(['On a scale of 1-100, how would you express this feeling?','How are you feeling right now?','Timestamp'],axis=1)
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    for colname in df.columns:
        if df[colname].dtypes=="object":
            df[colname] = le.fit_transform(df[colname])
    class_status = le.fit_transform(class_status)
    print(le.classes_)
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test = train_test_split(df,class_status,test_size = 0.2)
    from sklearn.tree import DecisionTreeClassifier
    dt = DecisionTreeClassifier()
    dt.fit(x_train,y_train)
    print("Training Acc: {}".format(dt.score(x_train,y_train)))
    print("Testing Acc: {}".format(dt.score(x_test,y_test)))
```

```python
simu = pd.read_csv("simu.csv")
simu = simu.sample(300)
simu.Timestamp = simu.Timestamp.apply(lambda x: datetime.datetime.strptime(x,"%m/%d/%Y %H:%M:%S") +datetime.timedelta(days = random.randrange(120,180)))
simu.to_csv("dataset_sim.csv",index=False)
train(simu)
```
