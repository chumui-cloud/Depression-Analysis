## Preprocessing

```python
info = []
for i in range((len(df))):
    info.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
info
```

**Encoding data to boolean values of array**

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr = TransactionEncoder()
t_info = tr.fit(info).transform(info)
```

```python
t_info
```

```python
df = pd.DataFrame(t_info,columns=tr.columns_) # create a dataframe
```

```python
df
```

**Using Apriori algo to create frequent itemsets**

```python
from mlxtend.frequent_patterns import apriori
frequent_items = apriori(df, min_support=0.60, use_colnames=True)
```

```python
frequent_items
```

## Association Rule Mining

```python
from mlxtend.frequent_patterns import association_rules
result = association_rules(frequent_items, metric='confidence', min_threshold=.60)
```
